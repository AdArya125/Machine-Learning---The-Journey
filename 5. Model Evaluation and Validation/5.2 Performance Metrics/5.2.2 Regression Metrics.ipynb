## Explanation of Various Regression Metrics

1. **Mean Absolute Error (MAE)**: The average of the absolute errors between predicted and actual values. It gives a measure of how far predictions are from actual values on average.
   $$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right| $$

2. **Mean Squared Error (MSE)**: The average of the squared errors between predicted and actual values. It gives more weight to larger errors.
   $$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2 $$

3. **Root Mean Squared Error (RMSE)**: The square root of the mean squared error. It provides an error metric in the same units as the response variable.
   $$ \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2} $$

4. **R-squared (Coefficient of Determination)**: A measure of how well the regression model fits the data. It represents the proportion of the variance in the dependent variable that is predictable from the independent variables.
   $$ R^2 = 1 - \frac{\sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2}{\sum_{i=1}^{n} \left( y_i - \bar{y} \right)^2} $$

### Scenarios and Examples of Using Each Regression Metric

1. **Mean Absolute Error (MAE)**:
   - **Scenario**: Suitable when all errors are equally important.
   - **Example**: In predicting house prices, MAE can provide a straightforward interpretation of the average prediction error.

2. **Mean Squared Error (MSE)**:
   - **Scenario**: Useful when larger errors are more significant.
   - **Example**: In financial forecasting, MSE penalizes larger errors more, which can be crucial in risk management.

3. **Root Mean Squared Error (RMSE)**:
   - **Scenario**: Suitable when the error metric needs to be in the same unit as the response variable.
   - **Example**: In weather forecasting, RMSE gives a clear understanding of the prediction error in terms of temperature units (e.g., degrees Celsius).

4. **R-squared**:
   - **Scenario**: Useful for understanding the proportion of the variance explained by the model.
   - **Example**: In a linear regression model for predicting sales based on advertising spend, R-squared indicates how well the model explains the variation in sales.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = fetch_california_housing()
X = data.data
y = data.target
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred,edgecolor='w')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--',lw=2,label="Line of Unity (x=y)")
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title('Measured vs Predicted Values')
plt.legend()
plt.show()
```

## Conclusion

Regression metrics are vital for evaluating the performance of regression models. Each metric provides different insights into the model's performance, and the choice of metric depends on the specific problem and context. MAE is suitable when all errors are equally important, MSE is useful when larger errors are more significant, RMSE offers an error metric in the same unit as the response variable, and R-squared provides a measure of how well the model explains the variance in the dependent variable. By understanding and applying these metrics, practitioners can make informed decisions about model selection and improvement.
