## Continuous Data Analysis Example

For continuous data analysis, we'll focus on performing a two-sample t-test
. This test compares the means of two independent groups to determine if they are significantly different from each other.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Generate example data: heights of two groups (in inches)
group1_heights = np.random.normal(loc=65, scale=3, size=30)  # Mean height of 65 inches, SD = 3
group2_heights = np.random.normal(loc=68, scale=3, size=30)  # Mean height of 68 inches, SD = 3

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(group1_heights, group2_heights)

# Output results
print("Two-Sample T-Test Results:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in heights between the two groups.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in heights between the two groups.")
l=len(group1_heights)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(np.arange(1,l+1),group1_heights)
plt.hlines(group1_heights.mean(),1,l,color='r',linestyle='-.',label=f'Mean {group1_heights.mean():.2f}')
plt.title("Group1")
plt.legend()
plt.subplot(1,2,2)
plt.plot(np.arange(1,l+1),group2_heights)
plt.hlines(group2_heights.mean(),1,l,color='r',linestyle='-.',label=f'Mean {group2_heights.mean():.2f}')
plt.title("Group2")
plt.legend()
plt.show()
```

### Categorical Data Analysis Example: Chi-Square Test for Independence

In this example, we'll use a Chi-Square test to analyze the relationship between gender and voting preference.
.


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Example data: Gender vs Voting Preference
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Preference': ['A', 'A', 'B', 'B'],
    'Count': [30, 20, 40, 10]
})

# Create contingency table
contingency_table = pd.pivot_table(data, values='Count', index='Gender', columns='Preference')

# Perform Chi-Square test for independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Chi-Square Test Results for Independence:")
print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"P-value: {p_val:.3f}")
print(f"Degrees of freedom: {dof}")
print("\nExpected frequencies:")
print(expected,'\n')

# Interpret results
alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: There is a significant association between Gender and Voting Preference.")
else:
    print("Fail to reject null hypothesis: There is no significant association between Gender and Voting Preference.")


print()
# Visualization
contingency_table.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Voting Preferences by Gender')
plt.legend(title='Preference')
plt.show()
```
