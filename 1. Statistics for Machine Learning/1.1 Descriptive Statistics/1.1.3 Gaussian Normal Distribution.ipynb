# 1.1.3 Gaussian Normal Distribution

## Introduction

The Gaussian Normal Distribution, commonly known as the normal distribution, is a fundamental concept in statistics and probability theory. It is widely used due to its unique properties and its prevalence in natural phenomena. This notebook will explore the properties of the normal distribution, provide visualizations, and discuss practical applications.

## Properties of the Gaussian Normal Distribution

The Gaussian Normal Distribution is characterized by its bell-shaped curve and is defined by two parameters:
- **Mean (μ)**: The central point of the distribution.
- **Standard Deviation (σ)**: Measures the spread or dispersion of the distribution.)


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```


# Plotting Gaussian Normal Distributions

```python
# Function to plot Gaussian Normal Distribution
def plot_gaussian(mu, sigma):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, mu, sigma)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'μ={mu}, σ={sigma}')
    plt.title('Gaussian Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_gaussian(0, 1)   # Standard normal distribution
```

```python
plot_gaussian(0, 2)   # Normal distribution with higher variance
```

```python
plot_gaussian(2, 1)   # Normal distribution with a different mean
```

# Making Comparisions between the graphs

```python
# Function to plot Gaussian Normal Distributions side by side
def plot_gaussians_side_by_side(distributions):
    fig, axs = plt.subplots(1, len(distributions), figsize=(18, 6), sharey=True)
    x = np.linspace(-10, 10, 1000)
    
    for ax, (mu, sigma) in zip(axs, distributions):
        y = norm.pdf(x, mu, sigma)
        ax.plot(x, y, label=f'μ={mu}, σ={sigma}', color='blue')
        ax.fill_between(x, y, alpha=0.2, color='blue')
        ax.axvline(mu, color='red', linestyle='--', label='Mean (μ)')
        ax.axvline(mu + sigma, color='green', linestyle='--', label='μ + σ')
        ax.axvline(mu - sigma, color='green', linestyle='--', label='μ - σ')
        ax.axvline(mu + 2*sigma, color='purple', linestyle='--', label='μ + 2σ')
        ax.axvline(mu - 2*sigma, color='purple', linestyle='--', label='μ - 2σ')
        ax.set_title(f'μ={mu}, σ={sigma}')
        ax.set_xlabel('X')
        ax.set_ylabel('Probability Density')
        ax.grid(True)
        ax.legend()
    
    plt.suptitle('Gaussian Normal Distributions', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# List of (mean, standard deviation) pairs for different Gaussian distributions
distributions = [(0, 1), (0, 2), (2, 1)]
# Plotting the Gaussian Normal Distributions side by side
plot_gaussians_side_by_side(distributions)
```
