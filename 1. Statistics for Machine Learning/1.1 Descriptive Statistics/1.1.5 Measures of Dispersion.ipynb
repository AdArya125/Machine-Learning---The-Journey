```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generate_data():
    # Normal distribution
    data_normal = np.random.normal(loc=0, scale=1, size=1000)
    
    # Skewed distribution
    data_skewed = np.random.exponential(scale=2, size=1000)
    
    # Bimodal distribution
    data_bimodal = np.concatenate([np.random.normal(loc=-2, scale=1, size=500), np.random.normal(loc=3, scale=1.5, size=500)])
    
    return data_normal, data_skewed, data_bimodal
```

```python
def calculate_dispersion(data):
    range = np.ptp(data)  # Peak to peak (range)
    variance = np.var(data)
    std_dev = np.std(data)
    return range, variance, std_dev
```

```python
def plot_data(data, title):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=30, kde=False, color='skyblue', label='Histogram', stat='density')
    
    # Plot KDE
    sns.kdeplot(data, color='b', linewidth=2, label='KDE')
    
    
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(data, color='orange')
    plt.title(f'{title} - Box Plot')
    plt.xlabel('Value')
    
    plt.tight_layout()
    plt.show()
```

```python
data_normal, data_skewed, data_bimodal = generate_data()
    
dispersion_normal = calculate_dispersion(data_normal)
print("Normal Distribution: Range = {:.2f}, Variance = {:.2f}, Standard Deviation = {:.2f}".format(*dispersion_normal))
plot_data(data_normal, 'Normal Distribution')
```

```python
dispersion_skewed = calculate_dispersion(data_skewed)
print("Skewed Distribution: Range = {:.2f}, Variance = {:.2f}, Standard Deviation = {:.2f}".format(*dispersion_skewed))
plot_data(data_skewed, 'Skewed Distribution')
```

```python
dispersion_bimodal = calculate_dispersion(data_bimodal)
print("Bimodal Distribution: Range = {:.2f}, Variance = {:.2f}, Standard Deviation = {:.2f}".format(*dispersion_bimodal))
plot_data(data_bimodal, 'Bimodal Distribution')
```
