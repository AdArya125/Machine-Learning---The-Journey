# 1.1.1 Continuous and Discrete Functions
## Introduction
In statistics and data analysis, it is crucial to understand the types of data you are working with. Data can generally be classified into two types: continuous and discrete.

- **Continuous Data**: Continuous data can take any value within a given range. It is often measured and can have infinite possibilities within that range. Examples include height, weight, temperature, and t
ime.
- **Discrete Data**: Discrete data can only take specific values, often counted in whole numbers. Examples include the number of children in a family, the number of cars in a parking lot, or the number of students in a classroom.

This notebook will explore continuous and discrete data through examples, visualizations, and statistical measures.


```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Setting Seaborn style for plots
sns.set(style="whitegrid")
```

```python
# Generating Continuous Data
# Example: Heights of people (in cm)
np.random.seed(40)
heights = np.random.normal(loc=170, scale=10, size=1000)
print(heights[:10])
```

```python
# Generating Discrete Data
# Example: Number of children in families
np.random.seed(40)
num_children = np.random.poisson(lam=2, size=1000)
print(num_children[:10])
```

```python
# Visualizing Continuous Data
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
sns.histplot(heights, kde=True, color='skyblue')
plt.title('Continuous Data: Heights of People')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.axvline(np.mean(heights), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {np.mean(heights):.2f}')
plt.legend()

# Visualizing Discrete Data
plt.subplot(1, 2, 2)
sns.histplot(num_children, discrete=True, color='orange')
plt.title('Discrete Data: Number of Children in Families')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')
plt.axvline(stats.mode(num_children).mode, color='red', linestyle='dashed', linewidth=2, label=f'Mode: {stats.mode(num_children).mode}')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Statistical Measures for Continuous Data
mean_height = np.mean(heights)
std_height = np.std(heights)
median_height = np.median(heights)
print(f"Continuous Data (Heights):")
print(f"Mean: {mean_height:.2f} cm")
print(f"Standard Deviation: {std_height:.2f} cm")
print(f"Median: {median_height:.2f} cm")
```

```python
# Statistical Measures for Discrete Data
mode_num_children = stats.mode(num_children).mode
mode_frequency = stats.mode(num_children).count
mean_num_children = np.mean(num_children)
std_num_children = np.std(num_children)
print(f"Discrete Data (Number of Children):")
print(f"Mode: {mode_num_children} (Frequency: {mode_frequency})")
print(f"Mean: {mean_num_children:.2f}")
print(f"Standard Deviation: {std_num_children:.2f}")
```

```python
# Additional Visualizations
# Boxplot for Continuous Data
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=heights, color='lightblue')
plt.title('Boxplot: Heights of People')

# Bar Plot for Discrete Data
plt.subplot(1, 2, 2)
unique, counts = np.unique(num_children, return_counts=True)
sns.barplot(x=unique, y=counts, hue=unique, dodge=False, palette='viridis', legend=False)
plt.title('Bar Plot: Number of Children in Families')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Practical Applications and Real-World Examples

Understanding the difference between continuous and discrete data is crucial in many real-world applications:

1. **Healthcare**: Continuous data such as blood pressure, cholesterol levels, and weight are monitored over time to track a patient's health. Discrete data, such as the number of hospital visits or the count of different types of medications taken, are also important for patient records.
   
2. **Economics**: Continuous data like GDP growth rates, inflation rates, and stock prices are analyzed to make economic forecasts. Discrete data, such as the number of unemployed individuals or the count of businesses in a sector, provide insights into economic conditions.

3. **Education**: Continuous data like test scores and attendance rates help in evaluating student performance. Discrete data such as the number of students in different grades or the number of courses offered are essential for administrative purposes.
