# t-Distributed Stochastic Neighbor Embedding (t-SNE) - Iris Dataset

t-Distributed Stochastic Neighbor Embedding (t-SNE) is a nonlinear dimensionality reduction technique that is particularly well-suited for visualizing high-dimensional data in a lower-dimensional space (typically 2D or 3D). Unlike PCA, which preserves global structure, t-SNE focuses on preserving local structure, making it effective at reflecting local relationships within the data.

### **Steps to perform t-SNE**:
1. **Standardize the Data**: Ensure the data has a mean of 0 and a standard deviation of 1.
2. **Apply t-SNE**: Use the t-SNE algorithm to reduce the data to 2 or 3 dimensions.
3. **Visualize the Results**: Plot the low-dimensional representation of the data.

In this example, we will apply t-SNE to the Iris dataset to visualize its structure in 2D.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply t-SNE to reduce to 2D
tsne = TSNE(n_components=2, random_state=44)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
# Create a DataFrame with the t-SNE components
tsne_df = pd.DataFrame(X_tsne, columns=['TSNE1', 'TSNE2'])
tsne_df['target'] = y
```

```python
# Plot the t-SNE results
plt.figure(figsize=(10, 6))
for target, color, label in zip(np.unique(y), ['r', 'g', 'b'], target_names):
    plt.scatter(tsne_df[tsne_df['target'] == target]['TSNE1'], 
                tsne_df[tsne_df['target'] == target]['TSNE2'], 
                color=color, label=label, alpha=0.7)
plt.title('Iris Data after t-SNE')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion

In this example, we applied t-SNE to the Iris dataset to reduce its dimensionality and visualize its structure in 2D. The results illustrate how t-SNE preserves local relationships within the data, making it easier to identify patterns and clusters.

### Key points:
- **Dimensionality Reduction**: t-SNE effectively reduced the data to 2 dimensions.
- **Local Structure Preservation**: t-SNE focuses on preserving local relationships, which can highlight clusters and patterns that may not be apparent with linear techniques like PCA.
- **Data Visualization**: We visualized the transformed data, showing distinct clusters corresponding to the different Iris species.

t-SNE is a powerful tool for visualizing high-dimensional data, providing insights into the underlying structure and relationships within the dataset.
